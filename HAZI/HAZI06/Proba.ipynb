```python
import pandas as pd
```

```python
def load_csv(csv_path: str) -> pd.DataFrame:
    return pd.read_csv(csv_path)

df = load_csv("2018_03.csv")
df
```

```python
def order_by_scheduled_time(df : pd.DataFrame) -> pd.DataFrame:
        return df.sort_values("scheduled_time", axis=0)

df = order_by_scheduled_time(df)
df
```

```python
def drop_columns_and_nan(df : pd.DataFrame) -> pd.DataFrame:
        df = df.drop(['from', 'to'], axis=1)
        df = df.dropna(axis=0)
        return df 

df = drop_columns_and_nan(df)
df
```

```python
def convert_date_to_day(df : pd.DataFrame) -> pd.DataFrame:
        df['day'] = pd.to_datetime(df['date']).dt.day_name()
        new_df = df.drop('date', axis=1)
        return new_df
        



df = convert_date_to_day(df)
df
```

```python

def convert_scheduled_time_to_part_of_the_day(df : pd.DataFrame) -> pd.DataFrame:
        convertDf = df
        convertDf['scheduled_time'] = pd.to_datetime(convertDf['scheduled_time'], format='%Y-%m-%d %H:%M:%S')

        convertDf['part_of_the_day'] = pd.cut(convertDf['scheduled_time'].dt.hour,
                                                bins=[0, 3.59, 7.59, 11.59, 15.59, 19.59, 23.59],
                                                labels=['late_night', 'early_morning', 'morning', 'afternoon', 'evening', 'night'])

        convertDf = convertDf.drop(["scheduled_time"], axis=1)
        return convertDf
df = convert_scheduled_time_to_part_of_the_day(df)
df
```

```python
import math as ma
def convert_delay(df : pd.DataFrame) -> pd.DataFrame:
    delayDf = df
    delayDf['delay'] = pd.cut(delayDf['delay_minutes'],
                              bins=[-1, 5, ma.inf],
                              labels=[0, 1])
    return delayDf

df = convert_delay(df)
df
```

```python
def drop_unnecessary_columns(df: pd.DataFrame) -> pd.DataFrame:
    dropColDf = df.drop(['train_id', 'actual_time', 'delay_minutes'], axis=1)
    return dropColDf
df = drop_unnecessary_columns(df)
df
```
